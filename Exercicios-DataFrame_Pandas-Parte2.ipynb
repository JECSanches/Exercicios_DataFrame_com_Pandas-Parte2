## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
porc_mat = df_mat/24
porc_por = df_por/18
porc_geo = df_geo/8
porc_ing = df_ing/8
porc_his = df_his/8
porc_fis = df_fis/12
porc_qui = df_qui/12
````

````python
# 2) 
total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
total_acertos
````

````python
# 3) 
porcentual_geral = total_acertos / 90 
# Sendo 90 o número total de questões
porcentual_geral
````

````python
# 4) 
fase_2 = total_acertos
fase_2['passou_fase2'] = 0

fase_2.loc[fase_2['Qt_acertos'] > 45, 'passou_fase2'] = 1

n_aprovados = fase_2['passou_fase2'].sum()
print(f'--> Um total de {n_aprovados} alunos foram aprovados para a segunda fase')

fase_2.head(10)
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# 1. Novo DataFrame com os alunos que tiraram zero.
zeros_mat, _ = df_mat[df_mat['Qt_acertos'] == 0].shape
zeros_fis, _ = df_fis[df_fis['Qt_acertos'] == 0].shape
zeros_qui, _ = df_qui[df_qui['Qt_acertos'] == 0].shape
print(f'--> {zeros_mat} alunos tiveram zero acertos em matemática, {zeros_fis} em física e {zeros_qui} em química')

total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui

df_desqualificados = total_acertos[(df_mat['Qt_acertos'] == 0) | 
                                   (df_fis['Qt_acertos'] == 0) | 
                                   (df_qui['Qt_acertos'] == 0)]
df_desqualificados.head()
````

````python
# 1. Resolvendo de uma outra forma (creio que seja mais interessante)
df_completo = pd.DataFrame(range(0,100), columns=['mat'])
df_completo['mat'] = df_mat
df_completo['por'] = df_por
df_completo['geo'] = df_geo
df_completo['ing'] = df_ing
df_completo['his'] = df_his
df_completo['fis'] = df_fis
df_completo['qui'] = df_qui
df_completo['total_acertos'] = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui

df_completo.index = range(1,101)

# Os alunos desqualificados estão agora indicados no DataFrame
df_completo['desqualificado'] = 0
df_completo.loc[(df_completo['mat'] == 0) | (df_completo['fis'] == 0) | (df_completo['qui'] == 0), 'desqualificado'] = 1

df_completo.head()
````

````python
# 1. Um DataFrama apenas com os alunos desqualificados (para o segundo método) pode ser obtido fazendo
df_alunos_desqualificados = df_completo.loc[df_completo['desqualificado'] == 1]
df_alunos_desqualificados.head()
````

````python
# 2. Número de alunos desqualificados
n_desqualificados, _ = df_desqualificados.shape
print(f'--> Um total de {n_desqualificados} alunos foram desqualificados por zerar em pelo menos uma das 3 disciplinas obrigatórias')

# 2. Com base no segundo método da questão 1.
n_desqualificados = df_completo['desqualificado'].sum()
print(f'--> Um total de {n_desqualificados} alunos foram desqualificados por zerar ou em matemárica ou física ou química.')
````

````python
# 3. Média em história e geografia dos alunos desqualificados
df_alunos_desqualificados = df_completo.loc[df_completo['desqualificado'] == 1]
media_geo = df_alunos_desqualificados['geo'].mean()
media_his = df_alunos_desqualificados['his'].mean()
print(f'--> A média em geográfia e história dos alunos desqualificados é {round(media_geo,2)} e {round(media_his,2)}, respectivamente')
````

````python
# 4. DataFrame com alunos qualificados para segunda fase
df_alunos_qualificados = df_completo[df_completo['desqualificado'] == 0]
df_alunos_qualificados.head()
````
